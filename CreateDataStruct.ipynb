```python
import pandas as pd
import numpy as np
```

```python
f1 = pd.read_csv('./dataFolder/inertManip_nd_tepfr_vert.csv', header = None)
f1.columns = ['Tracking_Error', '1', '2', 'concat']
```

```python
f1.head()
```

```python
m = []
k = []
for item in list(f1.concat):
    a, b = item[0:4], item[4:8]
    m.append(a)
    k.append(b)

f1['a'] = m
f1['t'] = k
f1.drop(columns = ['1', '2', 'concat'], inplace = True)
f1.head()
```

```python
f1.a.unique(), f1.t.unique()
```

```python
map_dict = {'m001': 0.1,
           'm010': 1,
           'm100': 10,
           'k001':0.1,
           'k010':1,
           'k100':10}

f1['abdo_mass'] = f1['a'].map(map_dict)
f1['torsional_spring'] = f1['t'].map(map_dict)
f1.drop(columns = ['a', 't'], inplace = True)
```

```python
f1['length_scale'] = [1]*len(f1)
f1['petiole_length'] = [0]*len(f1)
f1.head()
```

```python
for col in f1[['abdo_mass', 'torsional_spring', 'length_scale', 'petiole_length']]:
    print(f1[col].unique())
```

```python
f2 = pd.read_csv('./dataFolder/sishManip_nd_tepfr_vert.csv', header = None,)
f2.columns = ['Tracking_Error', '1', '2', '3', 'concat']
```

```python
f2['l'] = f2.concat.str.split('_', expand = True)[0]
f2['p'] = f2.concat.str.split('_', expand = True)[1]
```

```python
f2['l'].unique(), f2['p'].unique()
```

```python
map_dict = {'L05': 0.5,
           'L10': 1,
           'L20': 2,
           'E0': 0,
           'Ep2': 0.2,
           'Ep4': 0.4}
f2['length_scale'] = f2['l'].map(map_dict)
f2['petiole_length'] = f2['p'].map(map_dict)
f2['abdo_mass'] = [1]*len(f2)
f2['torsional_spring'] = [1]*len(f2)

f2.drop(columns = ['1', '2', '3', 'concat','l', 'p'], inplace = True)
f2.head()
```

```python
for col in f2[['abdo_mass', 'torsional_spring', 'length_scale', 'petiole_length']]:
    print(f2[col].unique())
```

```python
df = pd.concat([f1, f2], axis = 0, sort = True)
df.reset_index(drop = True, inplace = True)
```
