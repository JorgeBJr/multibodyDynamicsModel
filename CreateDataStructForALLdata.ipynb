```python
import pandas as pd
import numpy as np
```

## save the vertical dataset

```python
vertical = pd.read_csv('./dataFolder/allManipData_vert.csv')
vertical.drop(columns = ['LSF', 'motion', 'comboName'], inplace = True)
```

```python
vertical.rename(columns = {'m': 'abdominal_mass',
                          'K': 'spring_constant',
                          'LSF.1': 'length_scale',
                          'LE': 'petiole_length'},
               inplace = True)
vertical.head()
```

```python
for col in vertical[['abdominal_mass', 'spring_constant', 'length_scale', 'petiole_length']]:
    print(vertical[col].unique())
```

```python
s = vertical.duplicated()
subset = vertical[s]
```

```python
subset
```

```python
# finally save the data matrix in a useable form

final_dataStruct = vertical.drop(index = subset.index)
```

```python
len(vertical), len(final_dataStruct), 9*40 + 9*40 - 40
```

```python
final_dataStruct.to_csv('./dataFolder/AllData_vertically_oscillating.csv', index = False)
```
